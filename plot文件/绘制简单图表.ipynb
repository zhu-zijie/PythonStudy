```python
# 绘制简单的折线图
import matplotlib.pyplot as plt  # 再次导入matplotlib库

plt.figure(facecolor='w', figsize=(8, 5))  # 创建一个新图形，设置背景色为白色('w')，大小为8x5英寸

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

x = range(1, 15, 1)  # 创建x轴数据，从1到14，步长为1
y = range(1, 42, 3)  # 创建第一条线的y轴数据，从1到40，步长为3
plt.plot(x, y, color='r', linestyle='-.', marker='o', markerfacecolor='w')  # 折线颜色为红色，线型为点划线，标记为圆形，标记内部为白色

y = range(1, 57, 4)  # 创建第二条线的y轴数据，从1到56，步长为4
plt.plot(x, y, color='b', linestyle='--')  # 折线颜色为蓝色('b')，线型为虚线('--')

y = range(1, 71, 5)  # 创建第三条线的y轴数据，从1到70，步长为5
plt.plot(x, y, color='g', linestyle=':', marker='*')  # 折线颜色为绿色('g')，线型为点线(':')，标记为星形('*')

y = range(1, 85, 6)  # 创建第四条线的y轴数据，从1到84，步长为6
plt.plot(x, y, color='pink', linestyle='-', marker='o', mfc='w')  # 折线颜色为粉色，线型为实线，标记为圆形，mfc是markerfacecolor的缩写，表示标记内部为白色

plt.xlabel('X轴标签')  # 设置x轴标签
plt.ylabel('Y轴标签')  # 设置y轴标签
plt.show()  # 显示包含所有四条线的图表
```

```python
# 多柱状图
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 读取数据
df = pd.read_excel('产品销售统计.xlsx')

# 设置画布大小
plt.figure(figsize=(15, 10))
# x = df['产品名称']
x = np.array([i for i in range(8)])  # 将数组类型转为numpy数组
y1 = df['1月'] + df['2月'] + df['3月']
y2 = df['4月'] + df['5月'] + df['6月']
y3 = df['7月'] + df['8月'] + df['9月']
y4 = df['10月'] + df['11月'] + df['12月']

bar_width = 0.2  # 设置柱状图的宽度

# 设置y轴的标签
plt.ylabel('季度销售量')
plt.xlabel('产品名称')

# 设置标题
plt.title('2020年度产品销售统计', fontsize=15)

# 设置坐标轴的刻度
plt.xticks(x, df['产品名称'])

# 添加文本标签
for a, b in zip(x, y1):
    plt.text(a - bar_width * 1.5, b, b, ha='center', va='bottom', fontsize=8, color='y', alpha=0.9)
for a, b in zip(x, y2):
    plt.text(a - bar_width * 0.5, b, b, ha='center', va='bottom', fontsize=8, color='y', alpha=0.9)
for a, b in zip(x, y3):
    plt.text(a + bar_width * 0.5, b, b, ha='center', va='bottom', fontsize=8, color='y', alpha=0.9)
for a, b in zip(x, y4):
    plt.text(a + bar_width * 1.5, b, b, ha='center', va='bottom', fontsize=8, color='y', alpha=0.9)

# 绘制柱状图
plt.bar(x - bar_width * 1.5, y1, label='第一季度', color='r', alpha=0.8, width=bar_width)
plt.bar(x - bar_width * 0.5, y2, label='第二季度', color='b', alpha=0.8, width=bar_width)
plt.bar(x + bar_width * 0.5, y3, label='第三季度', color='g', alpha=0.8, width=bar_width)
plt.bar(x + bar_width * 1.5, y4, label='第四季度', color='y', alpha=0.8, width=bar_width)

# 设置图例
plt.legend(('第一季度', '第二季度', '第三季度', '第四季度'))

# 显示图表
plt.show()
```

```python
# 分裂饼图
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
df = pd.read_excel('产品销售统计.xlsx')

# 解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 设置画布大小
plt.figure(figsize=(12, 6))

# 设置标题
x = df['总量']
labels = df['产品名称']

# 绘制饼图，explode参数用于设置饼图的分裂程度,0表示不分裂
plt.pie(x, labels=labels, autopct='%1.1f%%', labeldistance=1.1, startangle=90, textprops={'fontsize': 12},
        explode=(0, 0, 0, 0.1, 0, 0, 0, 0), shadow=True)

# 设置x,y轴刻度的一致，保证饼图是一个正圆
plt.axis('equal')

# 标题
plt.title('2020年度产品销售统计')

# 显示图例
plt.legend(df['产品名称'], loc='upper right', frameon=False)

# 显示饼图
plt.show()
```
